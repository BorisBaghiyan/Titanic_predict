```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_train.drop('Transported', axis=1)
y = df_train['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
new_df1 = pd.concat([df_train[df_train['VIP'] == 1].head(300), df_train[df_train['VIP'] == 0].head(300)], ignore_index=True)
```

```python
new_df2 = pd.concat([df_test[df_test['VIP'] == 1].head(100), df_test[df_test['VIP'] == 0].head(100)], ignore_index=True)
```

```python
accuracy_list = []
k_list = []
for i in range(1, 50):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy_list.append(accuracy_score(y_test, y_pred))
    k_list.append(i)
```

```python
plt.plot(k_list, accuracy_list)
plt.xlim(10, 20)
```

```python
model = KNeighborsClassifier(n_neighbors=12)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors':[i for i in range(1, 15)],
    'metric': ["euclidean", "manhattan", "chebyshev", "minkowski", "mahalanobis"],
}
```
